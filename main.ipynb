```python
import configparser
import tweepy
import json
```

```python
config = configparser.ConfigParser()
config.read('config.ini') #file name
API_KEY = config['DEV']['API_KEY']
API_KEY_SECRET = config['DEV']['API_KEY_SECRET']
ACCESS_TOKEN = config['DEV']['ACCESS_TOKEN']
ACCESS_TOKEN_SECRET = config['DEV']['ACCESS_TOKEN_SECRET']
```

```python
auth = tweepy.OAuthHandler(API_KEY, API_KEY_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
api = tweepy.API(auth, wait_on_rate_limit=True, retry_count=10, 
                            retry_delay=5, retry_errors=5, timeout=60)
api.verify_credentials()
```

```python
count_limit = 500 
tweetsArr = []
count = 0
id = None
batch = 100
query = '#python'
```

```python
while count < count_limit:
    try:
        tweets = api.search_tweets(q=f"{query}", lang='es', tweet_mode='extended', max_id=id, count=batch)
        if tweets:
            print(f'{len(tweets)} tweets found and last id is {id}')
            for tweet in tweets:
                json_tweet = json.dumps(tweet._json)
                json_tweet = json.loads(json_tweet)
                tweetsArr.append(json_tweet)
                count += 1
                id = tweet.id
            if count>0:
                print(f"{count} tweets inserted")
        else:
            print('No tweets')
            break
    except Exception as e:
        print(e)
        break
```
